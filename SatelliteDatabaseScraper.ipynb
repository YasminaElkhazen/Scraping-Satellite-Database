```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup



def Scrape_Multiple_Pages(url, num_pages, k, merged_soup): 

# Switch WebDriver's focus to the target tab
    for i in range(k, num_pages + k):
        # Construct the XPath expression with the value of i: 
        xpath_expression = f'//a[contains(@href, "Page${i}")]'

        # Wait for the link element to be present
        next_link = WebDriverWait(driver, 35).until(
            EC.presence_of_element_located((By.XPATH, xpath_expression))
        )
        
        # Click the link
        next_link.click()
        print('Successfuly Scraped Page n', i)
        
        # Get the current page source after clicking the link
        page_source = driver.page_source
        
        # Create a BeautifulSoup object for the current page
        soup = BeautifulSoup(page_source, 'html.parser')
        print(soup)
        # Append the content of the current page to merged_soup
        merged_soup.extend(soup.contents)
        #Apparently there are length and time limits for soups so I printed out the content then copy pasted it into a .txt file
   
    return merged_soup

```

```python
url = 'https://www.heavens-above.com/Satellites.aspx'
num_pages = 40  #There's a time limit it only allowed 40 pages, So we're going to use a loop to scrape 1000 pages ;) 
driver = webdriver.Chrome()
driver.get(url)
merged_sou = BeautifulSoup("", "html.parser")
result_f=Scrape_Multiple_Pages(url,1,1,merged_sou)
for i in range(1, 250):    
    result_f.extend(Scrape_Multiple_Pages(url, num_pages, 40*i,merged_sou).contents)
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Read the HTML file with specified encoding
with open('C:\\Users\\ASUS\\Desktop\\satdb.txt', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all tables with class "standardTable"
tables = soup.find_all('table', class_='standardTable')

# List to store unique rows
unique_rows = []

# Process each table
for table in tables:
    # Extract table rows
    rows = table.find_all('tr')
        
    # Process each row
    for row in rows:
        # Extract text from each cell in the row and split by newline character
        row_values = [cell.text.strip() for cell in row.find_all(['th', 'td'])]
        
        # Append the row values to the list of unique rows
        unique_rows.append(row_values)

# Find the maximum number of columns in the extracted rows
max_columns = max(len(row) for row in unique_rows)

# Pad rows with fewer columns with empty strings to match the maximum number of columns
for row in unique_rows:
    while len(row) < max_columns:
        row.append('')

# Create a DataFrame from the list of unique rows
df = pd.DataFrame(unique_rows, columns=[f'Column {i+1}' for i in range(max_columns)])

# Display the DataFrame
print(df)
```
